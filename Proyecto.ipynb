# Jóvenes en el extranjero entre 18 y 24 años
Información de la población colombiana residente y registrada en las diferentes misiones consulares en el exterior que incluye el género, la edad, el nivel de estudios, la ocupación, lugar de residencia, el consulado de la circunscripción que lo atiende, entre otros.

Enlace de conjunto de datos: https://www.datos.gov.co/Estad-sticas-Nacionales/J-venes-en-el-extranjero-entre-18-y-24-a-os/4sd4-t4fe/about_data

Dataset suministrado  por el Ministerio de Relaciones Exteriores.

# Columnas del Datset
**País**: Indica el país de residencia de la persona. Cuando no se tienen datos registrados se presenta el texto

**Código ISO país**: ISO del país de residencia. Cuando no se tienen datos registrados se presenta el texto

**Oficina de registro**: Consulado donde la persona realizó su registro consular. Cuando no se tienen datos registrados se presenta el texto

**Grupo edad**: El grupo se basa en la edad entre 18 y 24 años Adulto joven.

**Edad (años)**: Edad de la persona. Cuando no se tienen datos registrados se el valor -1.

**Área Conocimiento**: Cuando no se tienen datos registrados se presenta el texto

**Sub Area Conocimiento**: Cuando no se tienen datos registrados se presenta el texto

**Nivel Académico**: Cuando no se tienen datos registrados se presenta el texto

**Estado civil**: Cuando no se tienen datos registrados se presenta el texto

**Género**: Cuando no se tienen datos registrados se presenta el texto

**Etnia de la persona**: Cuando no se tienen datos registrados se presenta el texto

**Estatura (CM)**: Estatura dada en centímetros, el valor -1 indica que no se encuentra una estatura registrada.

**Localización**: Localización del ciudadano georeferenciada según el país de residencia

**Cantidad de personas**: Cantidad de personas residentes en el exterior que cumplen con las condiciones demográficas

```python
duplicados = jov_en_extranj_df[jov_en_extranj_df.duplicated()]
print("Filas duplicadas:")
print(duplicados)
```

```python
Jov_EnExtranj_SinDup = jov_en_extranj_df.drop_duplicates()
```

```python
num_registros = Jov_EnExtranj_SinDup.shape[0]
print(f"Número de registros del dataset después de eliminar duplicados: {num_registros}")
```

```python
# Valores a filtrar de campo Localización encontradas con coordenadas (0.0, 0.0)
valores_a_filtrar = ['(0.0, 0.0)']
#, '-1', '139', '(0.0, 0.0)'

# Filtrar el DataFrame
df_coordenadas_cero = Jov_EnExtranj_SinDup[Jov_EnExtranj_SinDup['Localización'].isin(valores_a_filtrar)]
print(df_coordenadas_cero)
```

```python
# Eliminar los registros del DataFrame original que están en el DataFrame filtrado
Jov_EnExtranj_18a24_nvo = Jov_EnExtranj_SinDup.drop(df_coordenadas_cero.index)

# Mostrar el DataFrame después de eliminar los registros
print("\nDataFrame después de eliminar los registros filtrados:")
print(Jov_EnExtranj_18a24_nvo)
```

```python
num_registros = Jov_EnExtranj_18a24_nvo.shape[0]
print(f"El número de registros en el dataset después de eliminar registros filtrados con coordenadas cero: {num_registros}")
```

```python
# Creo función para dividir campo con coordenadas
def dividir_coordenadas(coordenada):
    valores = coordenada.strip("()").split(", ")
    return float(valores[0]), float(valores[1])
```

```python
# Aplicar la función y crear nuevas columnas
Jov_EnExtranj_18a24_nvo[['Coordenada1', 'Coordenada2']] = Jov_EnExtranj_18a24_nvo['Localización'].apply(lambda x: pd.Series(dividir_coordenadas(x)))
```
